```python
import os
import time
from multiprocessing import Pool
from event_selection import eventSelection
from event_selection import Quiet
from event_selection import get_efficiencies_df

signal_list  = [ "Tau_LQ_0250" , "LQ_LQ_0250" ]
signal_list += [ "Tau_LQ_0500" , "LQ_LQ_0500" ]
signal_list += [ "Tau_LQ_0750" , "LQ_LQ_0750" ]
signal_list += [ "Tau_LQ_1000" , "LQ_LQ_1000" ]
signal_list += [ "Tau_LQ_1250" , "LQ_LQ_1250" ]
signal_list += [ "Tau_LQ_1500" , "LQ_LQ_1500" ]
signal_list += [ "Tau_LQ_1750" , "LQ_LQ_1750" ]
signal_list += [ "Tau_LQ_2000" , "LQ_LQ_2000" ]
signal_list += [ "Tau_LQ_2250" , "LQ_LQ_2250" ]
signal_list += [ "Tau_LQ_2500" , "LQ_LQ_2500" ]

top_list=["ttbar","stop"]
diboson_list=["ww","wz","zz"]
v_jets_lists=["w_jets","z_jets"]

signal_list=top_list+v_jets_lists+diboson_list+signal_list

folder_out=os.path.join(
    os.getcwd(),
    "Selected_Events"
)

!rm -rf {folder_out}
!mkdir -p {folder_out}

def mapping(name):
    return eventSelection(name,folder_out)

with Quiet():
    start_time = time.time()
    with Pool(8) as p:
        goodEvents_dict=dict(p.map(mapping,signal_list))
    efficiencies=get_efficiencies_df(goodEvents_dict,folder_out)
    elapsed_seconds=(time.time() - start_time)
    hours=int(elapsed_seconds/3600)
    minutes=int((elapsed_seconds%3600)/60)
    seconts=int((elapsed_seconds%3600)%60)
    print("Done!")
    print(f"Time elapsed: {hours}h {minutes}m {seconts}s")
```

```python
for signal in goodEvents_dict.keys():
    print("%"*60)
    print(signal)
    print(goodEvents_dict[signal])
```

```python
for selection in efficiencies.keys():
    print("%"*80)
    print(selection)
    print(efficiencies[selection])
```
