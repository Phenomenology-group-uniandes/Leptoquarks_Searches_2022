```python
import pandas as pd
import numpy as np
import os

pd.set_option('display.float_format', '{:.2g}'.format) #Los DataFrame se imprimen con 2 cifras significativas
```

Definamos las rutas que necesitamos para leer los archivos cutflows

```python
Path_delphes_preselection = os.path.dirname(os.path.realpath('Table_LaTeX_JP.ipynb'))
Path_Leptoquarks_searches = os.path.dirname(Path_delphes_preselection)
Path_Selected_Events = f'{Path_delphes_preselection}/Selected_Events' #Carpeta donde están los csv preseleccionados
```

```python
channels = ['hadronic', 'semileptonic']
processes = ['b_b_tau_tau', 'b_tau_tau']
signals = ['LQ_LQ','Tau_LQ']
Masses = ['1000', '1250', '1500', '1750', '2000', '2250', '2500']
```

Leamoslos y construyamos una tabla con solo la sección eficaz y la eficiencia de cada proceso:

```python
cutflows = {'Betard_minus1': {}, 'Betard_0': {}}
```

```python
for process in processes:
    for channel in channels:
        cutflows['Betard_minus1'][f'{process}_{channel}'] = pd.read_csv(f'{Path_Selected_Events}/Cutflow_{process}_{channel}.csv', index_col= 0)
        if np.shape(cutflows['Betard_minus1'][f'{process}_{channel}'])[0] == 5: 
            cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].drop(labels=[1,2,4], axis=0)
            cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].replace(cutflows['Betard_minus1'][f'{process}_{channel}']['cut'][3], f'E_{process}_{channel}')
            cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].set_index('cut')
            cutflows['Betard_minus1'][f'{process}_{channel}'] = np.transpose(cutflows['Betard_minus1'][f'{process}_{channel}'])
            
        else: 
            cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].drop(labels=[1,2,3,5], axis=0)
            cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].replace(cutflows['Betard_minus1'][f'{process}_{channel}']['cut'][4], f'E_{process}_{channel}')
            cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].set_index('cut')
            cutflows['Betard_minus1'][f'{process}_{channel}'] = np.transpose(cutflows['Betard_minus1'][f'{process}_{channel}'])
            
        lista = list(cutflows['Betard_minus1'][f'{process}_{channel}'][f'E_{process}_{channel}'].keys())
        for i in range(len(lista)):
            if('wo_RHC' in lista[i]): 
                indice = i - 1
                break

        filas = [lista[i] for i in range(indice, len(lista))]
        cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].drop(labels = filas, axis=0)
        cutflows['Betard_minus1'][f'{process}_{channel}'] = cutflows['Betard_minus1'][f'{process}_{channel}'].sort_index(ascending=False)
```

```python
cutflows['Betard_minus1'][f'{process}_{channel}']
```

```python
for process in processes:
    for channel in channels:
        cutflows['Betard_0'][f'{process}_{channel}'] = pd.read_csv(f'{Path_Selected_Events}/Cutflow_{process}_{channel}.csv', index_col= 0)
        if np.shape(cutflows['Betard_0'][f'{process}_{channel}'])[0] == 5: 
            cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].drop(labels=[1,2,4], axis=0)
            cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].replace(cutflows['Betard_0'][f'{process}_{channel}']['cut'][3], f'E_{process}_{channel}')
            cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].set_index('cut')
            cutflows['Betard_0'][f'{process}_{channel}'] = np.transpose(cutflows['Betard_0'][f'{process}_{channel}'])
            
        else: 
            cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].drop(labels=[1,2,3,5], axis=0)
            cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].replace(cutflows['Betard_0'][f'{process}_{channel}']['cut'][4], f'E_{process}_{channel}')
            cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].set_index('cut')
            cutflows['Betard_0'][f'{process}_{channel}'] = np.transpose(cutflows['Betard_0'][f'{process}_{channel}'])
            
        lista = list(cutflows['Betard_0'][f'{process}_{channel}'][f'E_{process}_{channel}'].keys())
        for i in range(len(lista)):
            if('wo_RHC' in lista[i]): 
                indice = i - 1
                break

        filas = [lista[i] for i in range(5, indice - 2)]
        cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].drop(labels = filas, axis=0)
        cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].drop(labels = ['cut.1'], axis=0)
        
        cutflows['Betard_0'][f'{process}_{channel}'].index = [nombre.replace('_wo_RHC', '') for nombre in cutflows['Betard_0'][f'{process}_{channel}'].index]
        cutflows['Betard_0'][f'{process}_{channel}'] = cutflows['Betard_0'][f'{process}_{channel}'].sort_index(ascending=False)
```
