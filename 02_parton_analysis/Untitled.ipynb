```python
import os

from extract_kin import (
    readLHEF , 
    LHE_Loader , 
    make_histograms , 
    get_kinematics_row,
    get_event_by_child
)
```

```python
signal_list=["Tau_LQ_2500","Tau_LQ_M2500_gU0_25","Tau_LQ_M2500_gU0_50","Tau_LQ_M2500_gU1_00","Tau_LQ_M2500_gU2_00","Tau_LQ_M2500_gU3_00"]
```

```python
import pandas as pd
def get_singleLQ_histograms(name):
    results=[]
    jobs=LHE_Loader(name).Forest
    for run in jobs:
        j=1
        childs=readLHEF(run)
        for child in childs:
            event=get_event_by_child(child)  
            tau=event.getParticlesByIDs([-15,15])[0]
            tau.SetName("#tau")
            lq=event.getParticlesByIDs([-9000007,9000007])[0]
            lq.SetName("lq")
            results+=[get_kinematics_row(tau,lq)]
        break
    df=pd.DataFrame.from_records(results)
    histos=make_histograms(df)
    del df
    return name , histos
```

```python
histos=dict(map(get_singleLQ_histograms,signal_list))
```

```python
from ROOT import THStack , TLegend , kBlack, TCanvas
titles=histos[next(iter(histos))].keys()
colors = [3, 7, 6, kBlack, 2, 4, 8, 9, 1, 43, 97, 38]
linestyles = [1, 1, 1, 1, 1,1, 1, 10, 9, 8, 7, 6, 5, 4, 3]
folder_out=os.path.join(
    os.getcwd(),
    "Histos"
)
!rm -rf {folder_out}
!mkdir -p {folder_out}

for title in titles:
    print(title)
    hs=THStack("hs", title)
    c2 = TCanvas(title,"Histos",1024,860)
    x_1 = 0.75
    x_2 = 0.89
    legend = TLegend(x_1,0.65,x_2,0.89)
    for i, key in enumerate(histos.keys()):
        h=histos[key][title]
        h.SetDirectory(0)
        h.SetLineColor(colors[i])
        h.SetLineStyle(linestyles[i])
        h.SetLineWidth(2)
        try: 
            h.Scale(1.0/h.Integral())
        except:
            pass
        legend.AddEntry(h,key,"l")
        legend.SetBorderSize(0)
        h.Rebin(2)
        hs.Add(h)
    xlabel=h.GetXaxis().GetTitle() 
    hs.Draw("NOSTACK HIST")
    legend.Draw()
    hs.SetTitle(title)
    hs.GetXaxis().SetTitle(xlabel)
    path=os.path.join(folder_out,f"{title}.png")
    print(path)
    c2.SaveAs(path)
```
