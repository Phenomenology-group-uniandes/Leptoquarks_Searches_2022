### Código para generar las Paramcards

```python
import numpy as np
import shutil
import pandas as pd
import time

inicio = time.time()
```

```python
M = ["250","500","750","1000","1250","1500","1750","2000","2250","2500"]
g_U = ["0.25", "0.5", "1.0", "1.5","2.0", "2.5", "3.0"]
```

```python
ruta_modelo = "/disco1/SIMULACIONES/LeptoQuarks/UFOs/Mod_VLQ_UFO"
```

```python
linea_modelo = "import model " + ruta_modelo
linea_def1 = "define lq = vlq vlq~"
linea_def2 = "define ta = ta+ ta-"
linea_gen = "generate vlq > all all"
lineas_definiciones = [linea_modelo, linea_def1, linea_def2, linea_gen]

lineas_outputs = []
for i in range(len(M)):
    for j in range(len(g_U)):
        output_ij = "output /disco4/personal_folders/Joaquin/Paramcards_13/" + "M" + M[i].replace('.','_') + "_gU" + g_U[j].replace('.','_') + " -no-jpeg"
        lineas_outputs.append(output_ij)
        
lineas = np.concatenate((lineas_definiciones, lineas_outputs))

script = open ('folders_paramcards_M_gU_13.txt','w')
script.write(lineas[0])

for i in range(1,len(lineas)):
    script.write('\n' + lineas[i])
script.close()
```

```python
for i in range(len(M)):
    for j in range(len(g_U)):
        etiqueta = "M" + M[i].replace('.','_') + "_gU" + g_U[j].replace('.','_')
        
        nombre_script = 'generate_paramcards_' + etiqueta + '_13.txt'
        script = open (nombre_script,'w')
        
        launch = "launch /disco4/personal_folders/Joaquin/Paramcards_13/" + etiqueta + " -i"
        
        decay = "calculate_decay_widths"
        param_1 = "set nevents 10000"
        param_2 = "set MVLQ " + M[i] 
        param_3 = "set gU " + g_U[j]
        param_4 = "set betard33 -1"

        lineas = [launch,decay, param_1, param_2, param_3, param_4]
        
        script.write(lineas[0])
            
        for k in range(1,len(lineas)):    
            script.write('\n' + lineas[k])
            
        script.close()
        
        !./../../../Collider/MG5_aMC_v3_1_0/bin/mg5_aMC {nombre_script}
        !rm {nombre_script}
        
        source_folder = "/disco4/personal_folders/Joaquin/Paramcards_13/" + etiqueta
        source_file = "/disco4/personal_folders/Joaquin/Paramcards_13/" + etiqueta + "/Events/run_01/param_card.dat"
        target = "/disco4/personal_folders/Joaquin/Paramcards_13/param_card_" + etiqueta + ".dat"
        
        shutil.copy(source_file, target)
        shutil.rmtree(source_folder) #Borra la carpeta con los outputs de MadGraph
```

```python
fin = time.time()

segundos = fin-inicio
horas = int(segundos/3600)
segundos = segundos - horas*3600
minutos = int(segundos/60)
segundos = segundos - minutos*60

mensaje = "El código se demoro: " + str(horas) + " horas, " + str(minutos) + " minutos y " + str(segundos)[:4] + " segundos."
print(mensaje)
```
